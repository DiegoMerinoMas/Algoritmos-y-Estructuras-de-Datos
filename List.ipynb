```python
%%writefile source.cpp
#include<iostream>
#include<cstdlib>
#include<ctime>
#include<functional>
using namespace std;
struct nodo {
    int dato;
    nodo* sig;
    nodo(int d) {
        this->dato = d;
        sig = NULL;
    }
};
class lista {
    nodo* inicio;
    size_t cantidad;
public:
    lista() {
        inicio = NULL;
        cantidad = 0;
    }
    nodo* principio() {
        return inicio;
    }
    void push_front(int e) {
        nodo* n = new nodo(e);
        if (cantidad == 0)
            inicio = n;
        else {
            n->sig = inicio;
            inicio = n;
        }
        ++cantidad;
    }
    void print() {
        nodo* aux = inicio;
        while (aux != NULL) {
            cout << aux->dato << " ";
            aux = aux->sig;
        }
        cout << endl;
    }
    //1-obtener objeto en posicion = at()
    void print_at(int pos) {
        if (cantidad == 0)
            cout << "La lista esta vacia" << endl;
        else {
            if (pos == 0)
                cout << inicio->dato << endl;
            else {
                nodo* aux = inicio;
                for (int i = 0; i < pos; i++)
                    aux = aux->sig;
                cout << aux->dato << endl;
            }
        }
    }
    //2-eliminar en posicion(n), 
    void delete_at(int pos) {
        if (cantidad == 0)
            cout << "La lista esta vacia" << endl;
        else {
            if (pos == 0) {
                delete_front();
            }
            else if (pos == cantidad - 1) {
                delete_back();
            }
            else {
                nodo* aux = inicio;
                for (int i = 0; i < pos - 1; i++)
                    aux = aux->sig;
                nodo* aux2 = aux->sig;
                aux->sig = aux2->sig;
                delete aux2;
                --cantidad;
            }
        }
    }
    //3-eliminar al inicio
    void delete_front() {
        if (cantidad == 0)
            cout << "La lista esta vacia" << endl;
        else {
            nodo* aux = inicio;
            inicio = inicio->sig;
            delete aux;
            --cantidad;
        }
    }
    //4-eliminar al final
    void delete_back() {
        if (cantidad == 0)
            cout << "La lista esta vacia" << endl;
        else {
            nodo* aux = inicio;
            while (aux->sig->sig != NULL)
                aux = aux->sig;
            delete aux->sig;
            aux->sig = NULL;
            --cantidad;
        }
    }
    //5-ordenar elementos(n2)
    void sort() {
        if (cantidad == 0)
            cout << "La lista esta vacia" << endl;
        else {
            for (int i = 0; i < cantidad - 1; i++) {
                nodo* aux = inicio;
                for (int j = 0; j < cantidad - i - 1; j++) {
                    if (aux->dato > aux->sig->dato) {
                        int aux2 = aux->dato;
                        aux->dato = aux->sig->dato;
                        aux->sig->dato = aux2;
                    }
                    aux = aux->sig;
                }
            }
        }
    }
    //6-buscar por valor(n)
    void search(int e) {
        if (cantidad == 0)
            cout << "Lista vacia" << endl;
        else {
            nodo* aux = inicio;
            int pos = 0;
            while (aux != NULL) {
                if (aux->dato != e)
                    pos++;
                else {
                    cout << endl << "Encontrado en la posicion: " << pos << endl;
                }
                aux = aux->sig;
                if (pos >= cantidad)
                    cout << "No se encontro el elemento." << endl;
            }
        }
    }
    //7-invertir lista(n2)
    void invert() {
        if (cantidad == 0)
            cout << "La lista esta vacia" << endl;
        else {
            nodo* aux = inicio;
            nodo* aux2 = inicio->sig;
            inicio->sig = NULL;
            while (aux2 != NULL) {
                nodo* aux3 = aux2->sig;
                aux2->sig = aux;
                aux = aux2;
                aux2 = aux3;
            }
            inicio = aux;
        }
    }
    //8-buscar solo elementos pares
    void search_even() {
        if (cantidad == 0)
            cout << "La lista esta vacia" << endl;
        else {
            nodo* n = inicio;
            while (n != NULL) {
                if (n->dato % 2 == 0)
                    cout << n->dato << " ";
                n = n->sig;
            }
        }
    }
    //9-listar con uso de lambdas
    void print_List_lambda() {
        auto printList = [](nodo* n) {
            while (n != NULL) {
                cout << n->dato << " ";
                n = n->sig;
            }
        };
        printList(inicio);
    }

    //10-listar de forma recursiva
    void print_recursive(nodo* n) {
        if (n != NULL) {
            cout << n->dato << " ";
            print_recursive(n->sig);
        }
    }
};
int main() {
    lista l; int numero;
    srand(time(NULL));
    cout << "Ingrese el numero de elementos que desea ingresar: ";
    cin >> numero;
    cout << "Desea ingresar los elementos de la lista manualmente? (1-si, 0-no): ";
    int opcion;
    cin >> opcion;
    if (opcion == 1) {
        for (int i = 0; i < numero; i++) {
            int e;
            cout << "Ingrese el elemento: ";
            cin >> e;
            l.push_front(e);
        }
    }
    if (opcion == 0) {
        cout << "Lista generada aleatoriamente" << endl;
        for (int i = 0; i < numero; i++) {
            int e = rand() % 100;
            l.push_front(e);
        }
    }
    cout << endl;
    cout << "Lista generada: " << endl;
    l.print();
    cout << endl;
    cout << endl;
    cout << "Desea eliminar algun elemento de la lista? (1-si, 0-no): ";
    cin >> opcion;
    if (opcion == 1) {
        cout << "Desea eliminar el elemento al inicio? (1-si, 0-no): ";
        cin >> opcion;
        if (opcion == 1) {
            l.delete_front();
            cout << "Lista: " << endl;
            l.print();
        }
        cout << endl;
        cout << "Desea eliminar el elemento al final? (1-si, 0-no): ";
        cin >> opcion;
        if (opcion == 1) {
            l.delete_back();
            cout << "Lista: " << endl;
            l.print();
        }
        cout << endl;
        cout << "Desea eliminar algun elemento en una posicion? (1-si, 0-no): ";
        cin >> opcion;
        if (opcion == 1) {
            int pos;
            cout << "Ingrese la posicion: ";
            cin >> pos;
            l.delete_at(pos);
            cout << "Lista: " << endl;
            l.print();
        }
    }
    cout << endl;
    cout << "Desea ordenar los elementos de la lista? (1-si, 0-no): ";
    cin >> opcion;
    if (opcion == 1) {
        l.sort();
        cout << "Lista ordenada: " << endl;
        l.print();
    }
    cout << endl;
    cout << "Desea buscar un elemento en la lista? (1-si, 0-no): ";
    cin >> opcion;
    if (opcion == 1) {
        int e;
        cout << "Ingrese el elemento a buscar: ";
        cin >> e;
        l.search(e);
    }
    cout << endl;
    cout << "Desea invertir la lista? (1-si, 0-no): ";
    cin >> opcion;
    if (opcion == 1) {
        l.invert();
        cout << "Lista invertida: " << endl;
        l.print();
    }
    cout << endl;
    cout << "Desea buscar solo los elementos pares? (1-si, 0-no): ";
    cin >> opcion;
    if (opcion == 1) {
        l.search_even();
    }
    cout << endl;
    cout << "Desea listar con uso de lambdas? (1-si, 0-no): ";
    cin >> opcion;
    if (opcion == 1) {
        cout << "Lista con uso de lambdas: " << endl;
        l.print_List_lambda();
    }
    cout << endl;
    cout << "Desea listar con uso de recursividad? (1-si, 0-no): ";
    cin >> opcion;
    if (opcion == 1) {
        cout << "Lista con uso de recursividad: " << endl;
        l.print_recursive(l.principio());
    }
    return 0;
}
```
